```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set(style='whitegrid')
plt.rcParams['figure.figsize'] = (10, 6)
```

# Create sample sales dataset

```python
data = {
    'Product': ['iPhone 15', 'Samsung QLED 65"', 'Dyson Air Purifier', 'KitchenAid Mixer',
                'Sony WH-1000XM5', 'Amazon Echo Dot', 'Google Nest Hub'],
    'Category': ['Smartphones', 'Televisions', 'Kitchen Appliances', 'Kitchen Appliances',
                 'Audio Accessories', 'Smart Home', 'Smart Home'],
    'Units Sold': [1250, 430, 670, 580, 900, 310, 275],
    'Price': [250, 450, 200, 150, 80, 50, 60],
    'Returns (%)': [1.5, 2.0, 1.8, 1.0, 1.2, 3.5, 9.1]
}

df = pd.DataFrame(data)
df['Revenue'] = df['Units Sold'] * df['Price']
df
```

# Revenue by product

```python
top_products = df.sort_values(by='Revenue', ascending=False)
top_products
```

# Category-wise sales

```python
category_sales = df.groupby('Category')['Revenue'].sum().sort_values(ascending=False)
category_sales
```

# Monthly revenue simulation

```python
monthly_revenue = pd.DataFrame({
    'Month': ['January', 'February', 'March'],
    'Revenue': [445200, 508400, 594200]
})

monthly_revenue['Growth Rate (%)'] = monthly_revenue['Revenue'].pct_change().fillna(0) * 100

monthly_revenue
```

# Customer demographics (simulated)

```python
customer_demo = pd.DataFrame({
    'Age Group': ['18-24', '25-34', '35-44', '45-54'],
    'Percentage': [12, 42, 28, 18]
})
```

#  Visualization - Monthly Revenue Trend

```python
plt.plot(monthly_revenue['Month'], monthly_revenue['Revenue'], marker='o', color='blue')
plt.title('📈 Monthly Revenue Trend - Q1 2025')
plt.xlabel('Month')
plt.ylabel('Revenue ($)')
plt.grid(True)
plt.show()
```

# Visualization - Category-wise Sales

```python
sns.barplot(x=category_sales.values, y=category_sales.index, palette='viridis')
plt.title('📊 Revenue by Product Category')
plt.xlabel('Revenue ($)')
plt.ylabel('Category')
plt.show()
```

# Top 5 Products by Revenue

```python
sns.barplot(x=top_products['Revenue'][:5], y=top_products['Product'][:5], palette='magma')
plt.title('🏆 Top 5 Products by Revenue')
plt.xlabel('Revenue ($)')
plt.ylabel('Product')
plt.show()
```

# Customer Demographics

```python
sns.barplot(x=customer_demo['Age Group'], y=customer_demo['Percentage'], palette='coolwarm')
plt.title('🧑 Customer Age Distribution')
plt.ylabel('Percentage (%)')
plt.xlabel('Age Group')
plt.show()
```
